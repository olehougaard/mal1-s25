```python
import matplotlib.pyplot as plt
import numpy as np

def logistic(x):
  return 1 / (1 + np.exp(-x))

lim = 2.5
t = np.linspace(-lim, lim, 100)
sig = logistic(t)
sig_prime = np.exp(-t)/np.square(1 + np.exp(-t))
tanh = np.tanh(t)
swish = t * logistic(t)

ReLU = np.vectorize(lambda x: max(x, 0.))    
relu = ReLU(t)

top = lim
bot = -.1 * top

plt.figure(figsize=(16, 6))
plt.plot([-lim, lim], [0, 0], "k-")
plt.plot([-lim, lim], [.5 * top, .5 * top], "k:")
plt.plot([-lim, lim], [top, top], "k:")
plt.plot([0, 0], [1.1 * bot, 1.1 * top], "k-")
#plt.plot(t, sig, "b-", linewidth=2, label=r"$\sigma(t) = \dfrac{1}{1 + e^{-t}}$")
# plt.plot(t, sig_prime, "y-", linewidth=2, label=r"$\sigma'(t) = \sigma(t)*(1 - \sigma(t))$")
#plt.plot([-lim, 0], [0, 0], "r-", linewidth=2, label=r"$heaviside(t) = round(\sigma(t))$")
#plt.plot([0, 0], [0, 1], "r-", linewidth=2)
#plt.plot([0, lim], [1, 1], "r-", linewidth=2)
plt.plot(t, relu, "r-", linewidth=2, label="$ReLU(t) = max(t, 0) = \int_0^t heaviside(t)$")
#plt.plot(t, tanh, "g-", linewidth=2, label="$tanh(t)$")
#plt.plot(t, swish, "y-", linewidth=2, label=r"$Swish_\beta(t) = t\sigma(\beta t)$")
plt.xlabel("t")
plt.legend(loc="upper left")
plt.axis([-lim, lim, 1.1 * bot, 1.1 * top])
plt.gca().set_yticks([0, 0.25 * top, 0.5 * top, 0.75 * top, top])
plt.grid()
plt.show()
```
